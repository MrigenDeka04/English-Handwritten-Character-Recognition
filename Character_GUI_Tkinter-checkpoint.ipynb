# GUI for Character Recognition
Tkinter-based GUI for drawing and recognizing characters.

```python
import tkinter as tk
from tkinter import *
import numpy as np
import tensorflow as tf
from PIL import Image, ImageOps
import cv2
```

```python
MODEL_PATH = r'E:\JEC Internship\models\character_model.h5'
model = tf.keras.models.load_model(MODEL_PATH)
```

```python
class App:
    def __init__(self, root):
        self.root = root
        self.root.title('Character Recognizer')
        self.canvas = tk.Canvas(root, width=200, height=200, bg='white')
        self.canvas.pack()
        self.canvas.bind('<B1-Motion>', self.draw)
        
        self.button_predict = tk.Button(root, text='Recognize', command=self.predict)
        self.button_predict.pack()
        self.button_clear = tk.Button(root, text='Clear', command=self.clear)
        self.button_clear.pack()
        self.label_result = tk.Label(root, text='Draw a character')
        self.label_result.pack()
        
    def draw(self, event):
        x, y = event.x, event.y
        r = 8
        self.canvas.create_oval(x-r, y-r, x+r, y+r, fill='black')
        
    def clear(self):
        self.canvas.delete('all')
        self.label_result.config(text='Draw a character')
        
    def predict(self):
        self.canvas.postscript(file='temp.ps')
        img = Image.open('temp.ps').convert('L')
        img = img.resize((64,64))
        img = ImageOps.invert(img)
        img = np.array(img)/255.0
        img = img.reshape(1,64,64,1)
        pred = model.predict(img)
        char_index = np.argmax(pred)
        self.label_result.config(text=f'Predicted: {char_index}')

root = tk.Tk()
app = App(root)
root.mainloop()
```
